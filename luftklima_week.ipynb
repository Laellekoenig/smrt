```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import datetime as dt
```

```python
df = pd.read_csv("datasets/luftklima_avg_by_date.csv")
```

```python
df[df["Anzahl Stationen"] >= 20].head(10)
```

```python
juni = df[(df.Datum >= "2020-06-01") & (df.Datum <= "2020-06-31")]
```

```python
len(juni)
```

```python
juni.head()
```

```python
date = juni.Datum
temp = juni.Lufttemperatur
```

```python
plt.figure(figsize=(10, 8))
plt.plot(date, temp)
plt.xticks(rotation=90, fontsize=8)
plt.title("Average Temperature by Day in June")
plt.xlabel("Date")
plt.ylabel("C")
plt.show()
```

```python
months = range(1, 11)
year = 2021
```

```python
for month in months:
    if month < 10:
        start = f"{year}-0{month}-01"
        end = f"{year}-0{month}-31"
    else:
        start = f"{year}-{month}-01"
        end = f"{year}-{month}-31"

    df_month = df[(df.Datum >= start) & (df.Datum <= end)]
    
    days = df_month.Datum
    temps = df_month.Lufttemperatur

    plt.figure(figsize=(10, 8))
    plt.plot(days, temps)
    plt.xticks(rotation=90, fontsize=8)
    plt.title(f"Average Temperature by Day in 2021-{month}")
    plt.xlabel("Date")
    plt.ylabel("C")
    plt.ylim([-5, 30])
    plt.show()      
```
