```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pd.read_csv("datasets/verkehrsz√§hlungen_total_per_date.csv")
```

```python
df.head()
```

```python
df.Date = df.Date.apply(lambda x: dt.datetime.strptime(x, "%d.%m.%Y"))
```

```python
df.head()
```

```python
df = df[(df.Date > "2020-10-01") & (df.Date < "2021-10-01")]
```

```python
df.head()
```

```python
df = df.sort_values(by=["Date"])
```

```python
df.head()
```

```python
x = df.Date
```

```python
y = df.Total
```

```python
plt.figure(figsize=(10, 6), dpi=160)
plt.title("Total amount of vehicles counted in Basel\nFrom 2020-10-01 to 2021-10-01")
plt.xlabel("Date")
plt.ylabel("Vehicle count")
plt.plot(x, y)
plt.savefig("test.png")
```

```python
df2 = pd.read_csv("datasets/luftklima_avg_by_date.csv")
```

```python
df2.head()
```

```python
df2 = df2[(df2.Datum > "2020-10-01") & (df2.Datum < "2021-10-01")]
```

```python
df2.head()
```

```python
luft = pd.read_csv("datasets/luftklima_reformatted.csv")
```

```python
luft.head()
```

```python
luft = luft[(luft.Datum > "2020-10-01") & (luft.Datum < "2021-10-01")]
```

```python
luft.head()
```

```python
luft.Datum = luft.Datum.apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d"))
```

```python
df2.Datum = df2.Datum.apply(lambda x: dt.datetime.strptime(x, "%Y-%m-%d"))
```

```python
tmax = luft.groupby(by="Datum").max().reset_index()
```

```python
tmin = luft.groupby(by="Datum").min().reset_index()
```

```python
x1 = df2.Datum
y1 = df2.Lufttemperatur
x2 = tmax.Datum
y2 = tmax.Lufttemperatur
x3 = tmin.Datum
y3 = tmin.Lufttemperatur
```

```python
plt.figure(figsize=(10, 8), dpi=160)
plt.plot(x1, y1, label="average temperature", color="forestgreen")
plt.plot(x2, y2, label="maximum temperature", color="firebrick")
plt.plot(x3, y3, label="minimum temperature", color="mediumblue")
plt.legend()
plt.xlabel("Date")
plt.ylabel("¬∞C")
plt.title("Comparison of average, maximum and minimum Temperature in Basel")
plt.savefig("minmax.png")
```

```python
delta = y2 - y3
```

```python
plt.figure(figsize=(10, 8), dpi=160)
plt.plot(x2, delta)
```
