```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv("datasets/luftklima.csv", delimiter=";")
```

```python
df.head()
```

```python
len(df.Name)
```

```python
df_no_nan = df.dropna()
```

```python
len(df_no_nan)
```

```python
dff = df_no_nan
```

```python
dff = dff.drop("Stadtklima Basel Link", 1)
```

```python
dff = dff.drop("Unix Zeitstempel", 1)
```

```python
dff.head()
```

```python
dff["longitude"] = dff.Koordinaten.apply(lambda entry: entry.split(",")[0])
```

```python
dff["latitude"] = dff.Koordinaten.apply(lambda entry: entry.split(",")[1])
```

```python
dff.head()
```

```python
dff = dff.drop("Koordinaten", 1)
```

```python
dff.head()
```

```python
dff.Zeitstempel = dff.Zeitstempel.apply(lambda entry: dt.datetime.fromisoformat(entry))
```

```python
dff.head()
```

```python
dff["year"] = dff.Zeitstempel.apply(lambda entry: entry.year)
```

```python
dff["month"] = dff.Zeitstempel.apply(lambda entry: entry.month)
```

```python
dff["day"] = dff.Zeitstempel.apply(lambda entry: entry.day)
```

```python
dff["hour"] = dff.Zeitstempel.apply(lambda entry: entry.hour)
```

```python
dff["minute"] = dff.Zeitstempel.apply(lambda entry: entry.minute)
```

```python
dff.head()
```

```python
dff.to_csv("datasets/luftklima_reformatted.csv", index=False)
```
