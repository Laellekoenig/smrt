```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("datasets/verkehrszählungen.csv", delimiter=";")
```

```python
df.head()
```

```python
dff = df.drop("ZST_NR", 1)
dff = dff.drop("SiteName", 1)
dff = dff.drop("DateTimeFrom", 1)
dff = dff.drop("DateTimeTo", 1)
dff = dff.drop("DirectionName", 1)
dff = dff.drop("LaneCode", 1)
dff = dff.drop("LaneName", 1)
dff = dff.drop("Weekday", 1)
dff = dff.drop("DayOfYear", 1)
dff = dff.drop("ValuesEdited", 1)
dff = dff.drop("Zst_id", 1)
```

```python
dff.head()
```

```python
dff.columns
```

```python
len(dff)
```

```python
# drop all values that are not approved
app = dff[dff.ValuesApproved == 1]
```

```python
len(app)
```

```python
# and now drop approved col
app = app.drop("ValuesApproved", 1)
```

```python
#format koords:
app["latitude"] = app["Geo Point"].apply(lambda x: float(x.split(",")[0]))
app["longitude"] = app["Geo Point"].apply(lambda x: float(x.split(",")[1]))
```

```python
app.head()
```

```python
# drop na values
app = app.dropna()
```

```python
len(app)
```

```python
# some info about set
maxLat = app.latitude.max()
minLat = app.latitude.min()
maxLon = app.longitude.max()
minLon = app.longitude.min()
minYear = app.Date.min()
maxYear = app.Date.max()

print(f"Minimum Latitude: {minLat}")
print(f"Maximim Latitude: {maxLat}")
print(f"Minimum Longitude: {minLon}")
print(f"Maximim Longitude: {maxLon}")
print(f"Minimum Date: {minYear}")
print(f"Maximum Date: {maxYear}")
```

```python
app.to_csv("datasets/verkerszählungen_reformatted.csv", index=False)
```
