```python
# remove entries with no coordinates
nn = df[(df.longitude != 0) & (df.latitude != 0)]
```

```python
s = []
for long, lat in zip(nn.longitude, nn.latitude):
    s.append(str(long) + "," + str(lat))
```

```python
s[0]
```

```python
nn["koords"] = s
```

```python
nn.head()
```

```python
uni = np.unique(s)
```

```python
len(uni)
```

```python
longs = []
lats = []
for item in uni:
    split = item.split(",")
    #delete invalid
    la = float(split[0])
    lo = float(split[1])
    
    if (np.floor(la) == 47.0 and np.floor(lo) == 7.0):
        longs.append(lo)
        lats.append(la)
```

```python
BBox = (min(longs), max(longs), min(lats), max(lats))
```

```python
BBox
```

```python
basel = plt.imread("maps/basel_smartklima.png")
```

```python
plt.figure(figsize=(5, 10))
plt.scatter(longs, lats, marker="o", s=300, zorder=1, alpha=0.7)
plt.title("Every Smartklima Measuring Station in Basel")
plt.xlim(BBox[0], BBox[1])
plt.ylim(BBox[2], BBox[3])

plt.imshow(basel, zorder=0, extent = BBox)
plt.show()
```
