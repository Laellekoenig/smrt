```python
df.head()
```

```python
df = df[["Station-ID", "Name", "Zeitstempel", "Datum", "Stunde", "Minute", "Koordinaten", "Breitengrad", "Längengrad", "Lufttemperatur", "Regen in 1 h", "Regen in 24 h", "Regen in 48 h"]]
```

```python
df.head()
```

```python
df_dates = df.groupby(df.Datum).mean()
```

```python
df_dates.head()
```

```python
df_dates = df_dates.drop(["Stunde", "Minute", "Breitengrad", "Längengrad"], 1)
```

```python
df_dates.head()
```

```python
df_dates = df_dates.sort_values(by=["Datum"])
```

```python
df_dates.head()
```

```python
df_dates = df_dates.reset_index()
```

```python
df_dates.head()
```

```python
df.groupby(df.Datum).size()
```

```python
df.groupby(df.Datum).size().reset_index(name="Count").Count
```

```python
station_count = []
for date in df.Datum.unique():
    date_entries = df[df.Datum == date]
    count = date_entries["Station-ID"].unique().size
    station_count.append(count)
```

```python
len(station_count) == df.Datum.unique().size
```

```python
station_count = np.array(station_count)
```

```python
df_dates["Anzahl Stationen"] = station_count
```

```python
df_dates["Anzahl Messungen"] = df.groupby(df.Datum).size().reset_index(name="Count").Count
```

```python
df_dates.head()
```

```python
df_dates.to_csv("datasets/luftklima_avg_by_date.csv")
```

```python
df.to_csv("datasets/luftklima_reformatted.csv")
```

```python
dat = df_dates.Datum
```

```python
temp = df_dates.Lufttemperatur
```

```python
plt.figure(figsize=(15, 10))
plt.plot(dat, temp)
plt.title("Average Temperature measured by Day")
plt.xlabel("Date")
plt.ylabel("°C")
```

```python
count = df_dates["Anzahl Messungen"]
```

```python
plt.figure(figsize=(15, 10))
plt.plot(dat, count)
plt.title("Number of individual Measurments per Day")
plt.xlabel("Date")
plt.ylabel("Number of Measurments")
```

```python
s_count = df_dates["Anzahl Stationen"]
```

```python
plt.figure(figsize=(15, 10))
plt.plot(dat, s_count)
plt.title("Number of unique Stations per Day")
plt.xlabel("Date")
plt.ylabel("Number of unique Stations")
```

```python
avg_per_station = count / s_count
```

```python
plt.figure(figsize=(15, 10))
plt.plot(dat, avg_per_station)
plt.title("Average Number of Measurments per Station per Day")
plt.xlabel("Date")
plt.ylabel("Average Number of Measurments")
```
